```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Charger une image en niveau de gris
img = cv2.imread(r'C:/M2/vision/photos/dogGris.jpg', cv2.IMREAD_GRAYSCALE)

# 2. Vrifier si le chargement s'est fait correctement
if img is None:
    print("Erreur: L'image n'a pas pu tre charge.")
else:
    # 3. Crer une nouvelle image du mme taille que img
    imgResultat = np.zeros_like(img)

    # 4. Inverser les valeurs des pixels (255 - niveau de gris de img)
    imgResultat = 255 - img

    # 5. Afficher img et imgResultat
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title("Image Originale (Gris)")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(imgResultat, cmap='gray')
    plt.title("Image Inverse (Gris)")
    plt.axis('off')

    plt.show()



# Mme chose avec une autre photo
# 1.Charger une image couleur
img_color = cv2.imread(r'C:/M2/vision/photos/cat.jpg', cv2.IMREAD_COLOR)

# 2. Vwrifier si le chargement s'est fait correctement
if img_color is None:
    print("Erreur: L'image couleur n'a pas pu tre charge.")
else:
    # 3. Crer une nouvelle image du mme taille que img_color
    #imgResultat_color = np.zeros_like(img_color)

    # 4. Inverser les valeurs des pixels pour chaque canal de couleur
    imgResultat_color = 255 - img_color

    # 5. Afficher img_color et imgResultat_color
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
    plt.title("Image Originale (Couleur)")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(imgResultat_color, cv2.COLOR_BGR2RGB))
    plt.title("Image Inverse (Couleur)")
    plt.axis('off')

    plt.show()
```

```python
import cv2

# 1. Charger une vido
cap = cv2.VideoCapture(r'C:/M2/vision/videos/dog.mp4')

# Vrifier si la vido a t ouverte correctement
if not cap.isOpened():
    print("Erreur: Impossible de lire la vido")
    exit()

while True:
    # 2. Lire la vido frame par frame
    ret, frame = cap.read()
    
    # Si on arrive  la fin de la vido, sortir de la boucle
    if not ret:
        print("Fin de la vido ou Erreur de lecture")
        break
    
    # 3. Manipulation : convertir en niveaux de gris
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # 4. Manipulation : inverser les couleurs
    inverted_frame = 255 - gray_frame
    
    # 5. Afficher la vido originale et la vido modifie
    cv2.imshow('Vido Originale', frame)
    cv2.imshow('Vido en Niveaux de Gris Inverse', inverted_frame)
    
    # 6. Sortir en appuyant sur la touche 'q'
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

# Librer les ressources et fermer les fentres
cap.release()
cv2.destroyAllWindows()
```
