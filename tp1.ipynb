```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Charger une image en niveau de gris
img = cv2.imread(r'C:/M2/vision/photos/dogGris.jpg', cv2.IMREAD_GRAYSCALE)

# 2. Vérifier si le chargement s'est fait correctement
if img is None:
    print("Erreur: L'image n'a pas pu être chargée.")
else:
    # 3. Créer une nouvelle image du même taille que img
    imgResultat = np.zeros_like(img)

    # 4. Inverser les valeurs des pixels (255 - niveau de gris de img)
    imgResultat = 255 - img

    # 5. Afficher img et imgResultat
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title("Image Originale (Gris)")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(imgResultat, cmap='gray')
    plt.title("Image Inversée (Gris)")
    plt.axis('off')

    plt.show()



# Même chose avec une autre photo
# 1.Charger une image couleur
img_color = cv2.imread(r'C:/M2/vision/photos/cat.jpg', cv2.IMREAD_COLOR)

# 2. Vwrifier si le chargement s'est fait correctement
if img_color is None:
    print("Erreur: L'image couleur n'a pas pu être chargée.")
else:
    # 3. Créer une nouvelle image du même taille que img_color
    #imgResultat_color = np.zeros_like(img_color)

    # 4. Inverser les valeurs des pixels pour chaque canal de couleur
    imgResultat_color = 255 - img_color

    # 5. Afficher img_color et imgResultat_color
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
    plt.title("Image Originale (Couleur)")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(imgResultat_color, cv2.COLOR_BGR2RGB))
    plt.title("Image Inversée (Couleur)")
    plt.axis('off')

    plt.show()
```

```python
import cv2

# 1. Charger une vidéo
cap = cv2.VideoCapture(r'C:/M2/vision/videos/dog.mp4')

# Vérifier si la vidéo a été ouverte correctement
if not cap.isOpened():
    print("Erreur: Impossible de lire la vidéo")
    exit()

while True:
    # 2. Lire la vidéo frame par frame
    ret, frame = cap.read()
    
    # Si on arrive à la fin de la vidéo, sortir de la boucle
    if not ret:
        print("Fin de la vidéo ou Erreur de lecture")
        break
    
    # 3. Manipulation : convertir en niveaux de gris
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # 4. Manipulation : inverser les couleurs
    inverted_frame = 255 - gray_frame
    
    # 5. Afficher la vidéo originale et la vidéo modifiée
    cv2.imshow('Vidéo Originale', frame)
    cv2.imshow('Vidéo en Niveaux de Gris Inversée', inverted_frame)
    
    # 6. Sortir en appuyant sur la touche 'q'
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

# Libérer les ressources et fermer les fenêtres
cap.release()
cv2.destroyAllWindows()
```
